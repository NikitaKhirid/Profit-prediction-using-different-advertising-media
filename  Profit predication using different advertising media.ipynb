# What Is Multiple Linear Regression (MLR)?

Multiple linear regression (MLR), also known simply as multiple regression, is a statistical technique that uses several explanatory variables to predict the outcome of a response variable. The goal of multiple linear regression is to model the linear relationship between the explanatory (independent) variables and response (dependent) variables.

# Formula of Multiple Linear Regression

- Multiple linear regression models are defined by the equation

**Y=β0+β1X1+β2X2+⋯+βpXp+ϵ**

It is similar than the equation of simple linear regression, except that there is more than one independent variables (X1,X2,…,Xp).

- Estimation of the parameters 
β0,…,βp by the method of least squares is based on the same principle as that of simple linear regression, but applied to p-dimensions. It is thus no longer a question of finding the best line (the one which passes closest to the pairs of points (yi,xi)), but finding the p-dimensional plane which passes closest to the coordinate points (yi,xi1,…,xip).

# 5. Handling Outliers 

**1 — What is an Outlier?**

An outlier is a data point in a data set that is distant from all other observations.

**2 — How can we Identify an outlier?**

- Using Box plots

- Using Scatter plot

- Using Z score

I've used Box Plots above in data visualization step to detect outliers.


```python
# Detection of outliers
Q1=np.percentile(df["Profit"],25,interpolation='midpoint')
Q3=np.percentile(df["Profit"],75,interpolation='midpoint')
IQR=Q3-Q1
print("Old shape : ",df.shape)
# Upper bound
upper=np.where(df["Profit"]>=(Q3+1.5*IQR))
print("Upper bound will be : ",upper)
# Lower bound
lower=np.where(df["Profit"]<=(Q1-1.5*IQR))
print("Lower bound will be : ",lower)
```

```python
# Removing the outliers
df.drop(upper[0],inplace=True)
df.drop(lower[0],inplace=True)
print("New Shape : ",df.shape)
```

```python
sns.boxplot(data=df,orient="v",color="red")
plt.title("Boxplot")
plt.show()
```

### Conclusion:
WE successfully remove the outliers from our dataset

# 7. Encoding Categorical Data
Machine learning algorithms cannot work with categorical data directly, categorical data must be converted to number.

  1.Label Encoding
  
  2.One hot encoding
  
  3.Dummy variable trap

### Label encoding:
Refers to transforming the word labels into numerical form so that the algorithms can understand how to operate on them.

### One hot encoding:
A One hot encoding is a representation of categorical variable as binary vectors.It allows the representation of categorical data to be more expresive. This first requires that the categorical values be mapped to integer values, that is label encoding. Then, each integer value is represented as a binary vector that is all zero values except the index of the integer, which is marked with a 1.

### Dummy variable trap:
The Dummy variable trap is a scenario in which the independent variable are multicollinear, a scenario in which two or more variables are highly correlated in simple term one variable can be predicted from the others.

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct=ColumnTransformer(transformers=[("encoder",OneHotEncoder(),[3])],remainder="passthrough")
x=(ct.fit_transform(x))
```

# 8. TRAIN TEST SPLIT

- The train-test split is a technique for evaluating the performance of a machine learning algorithm.

- Train Dataset: Used to fit the machine learning model.
- Test Dataset: Used to evaluate the fit machine learning model.

- Common split percentages include:

Train: 80%, Test: 20%

Train: 67%, Test: 33%

Train: 50%, Test: 50%

I've used 80% train and 20% test

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train.shape,y_train.shape
```

```python
x_test.shape,y_test.shape
```

# 9. Build the Regression model

**a. Fit Best Model**

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

**b. Predict on testing data using that model**

```python
y_pred=regressor.predict(x_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))

y_pred_train=regressor.predict(x_train)
```

 **c. Performance Metrics**


```python
from sklearn.metrics import mean_squared_error
MSE=mean_squared_error(y_pred,y_test)
print("The mean square error(MSE): ",MSE)
```

```python
RMSE=np.sqrt(mean_squared_error(y_pred,y_test))
print("The root mean square error(RMSE): ",RMSE)
```

```python
from sklearn.metrics import mean_absolute_error
MAE=mean_absolute_error(y_pred,y_test)
print("The mean absolute error(MAE): ",MAE)
```

```python
from sklearn.metrics import mean_absolute_percentage_error
MAPE=mean_absolute_percentage_error(y_pred,y_test)
print("The mean absolute  percentage error(MAE): ",MAPE)
```

```python
from sklearn.metrics import r2_score
R_Squ=r2_score(y_pred,y_test)
print("The R_Square value or coefficient of determination: ",R_Squ)
```

```python
print('Train Score: ', regressor.score(x_train, y_train))  
print('Test Score: ', regressor.score(x_test, y_test))  
```
